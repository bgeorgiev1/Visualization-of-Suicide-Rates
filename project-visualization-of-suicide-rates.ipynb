```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", None, "display.max_columns", None)

df = pd.read_csv('who_suicide_statistics.csv')
```

```python
# Dropping the year 2016 due to a lack of data resulting in misleading diagrams

df.drop(df[df['year'] == 2016].index, inplace = True)
```

```python
# Creating a chart representing the general suicide rates for the period 

plt.figure(figsize=(10,5))
by_year = df.groupby(['year'])[['population','suicides_no']].sum().reset_index()
by_year['suicide_rate'] = by_year['suicides_no']*100000/by_year['population']
ax = sns.lineplot(x="year", y="suicide_rate",data=by_year)
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Rates by Year across globe')
plt.show()
```

```python
# A graph juxtaposing the suicide rates based on sex

plt.figure(figsize=(12,5))
by_year = df.groupby(['year','sex'])[['population','suicides_no']].sum().reset_index()
by_year['suicide_rate'] = by_year['suicides_no']*100000/by_year['population']
ax = sns.lineplot(x="year", y="suicide_rate",hue = 'sex',data=by_year,  legend="full")
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Rates by Sex across globe')
plt.show()
```

```python
# A chart showing the suicide rates by age group

plt.figure(figsize=(12,7))
by_year = df.groupby(['year','age'])[['population','suicides_no']].sum().reset_index()
by_year['suicide_rate'] = by_year['suicides_no']*100000/by_year['population']
ax = sns.lineplot(x="year", y="suicide_rate",hue = 'age',data=by_year,  legend="full")
ax.set(ylabel='Suicide Rate', xlabel='Year', title = 'Suicide Rates by age group across globe')
plt.show()
```
